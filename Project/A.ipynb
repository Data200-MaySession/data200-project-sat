```python
fig, axs = plt.subplots(2, 3, figsize=(10, 5))
plt1 = sns.boxplot(housing['median_house_value'], ax=axs[0,0])
plt2 = sns.boxplot(housing['housing_median_age'], ax=axs[0,1])
plt3 = sns.boxplot(housing['total_rooms'], ax=axs[0,2])
plt4 = sns.boxplot(housing['total_bedrooms'], ax=axs[1,0])
plt5 = sns.boxplot(housing['population'], ax=axs[1,1])
plt6 = sns.boxplot(housing['median_income'], ax=axs[1,2])

plt.tight_layout()
```

```python
plt.boxplot(housing.median_house_value)
Q1 = housing.median_house_value.quantile(0.25)
Q3 = housing.median_house_value.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.median_house_value >= Q1 - 1.5*IQR) & (housing.median_house_value <= Q3 + 1.5*IQR)]
```

```python

plt.boxplot(housing.total_rooms)
Q1 = housing.total_rooms.quantile(0.25)
Q3 = housing.total_rooms.quantile(0.75)
IQR = Q3 - Q1
housing = housing[(housing.total_rooms >= Q1 - 1.5*IQR) & (housing.total_rooms <= Q3 + 1.5*IQR)]
```

```python

fig, axs = plt.subplots(2, 3, figsize=(10, 5))
plt1 = sns.boxplot(housing['median_house_value'], ax=axs[0, 0])
plt2 = sns.boxplot(housing['total_rooms'], ax=axs[0, 1])
plt3 = sns.boxplot(housing['total_bedrooms'], ax=axs[0, 2])
plt4 = sns.boxplot(housing['households'], ax=axs[1, 0])
plt5 = sns.boxplot(housing['housing_median_age'], ax=axs[1, 1])
plt6 = sns.boxplot(housing['median_income'], ax=axs[1, 2])

plt.tight_layout()
```

```python
sns.pairplot(housing)
plt.show()
```

```python
plt.figure(figsize=(20, 12))

# Using ocean_proximity as our main categorical variable
plt.subplot(2, 3, 1)
sns.boxplot(x='ocean_proximity', y='median_house_value', data=housing)

# For other plots, we'll use numerical variables binned into categories
plt.subplot(2, 3, 2)
sns.boxplot(x=pd.cut(housing['total_rooms'], bins=3), y='median_house_value', data=housing)

plt.subplot(2, 3, 3)
sns.boxplot(x=pd.cut(housing['total_bedrooms'], bins=3), y='median_house_value', data=housing)

plt.subplot(2, 3, 4)
sns.boxplot(x=pd.cut(housing['population'], bins=3), y='median_house_value', data=housing)

plt.subplot(2, 3, 5)
sns.boxplot(x=pd.cut(housing['median_income'], bins=3), y='median_house_value', data=housing)

plt.subplot(2, 3, 6)
sns.boxplot(x=pd.cut(housing['housing_median_age'], bins=3), y='median_house_value', data=housing)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))


housing['high_income'] = np.where(housing['median_income'] > housing['median_income'].median(), "Yes", "No")

sns.boxplot(
    x='ocean_proximity',      
    y='median_house_value',   
    hue='high_income',        
    data=housing
)
plt.show()
```
